```python
import numpy as np

filename="final_count.txt"
with open(filename) as f:
    lines = f.read().splitlines()

count=[]
for line in lines:
    count_line = line.split()[3]
    count.append(np.log10(1+float(count_line)))
#     count.append(float(count_line))

count = np.array(count)
# hist=np.histogram(count,bins='auto')
print(np.max(count))
# import matplotlib.pyplot as plt 
# plt.figure()
# plt.plot(hist[0])

import matplotlib.pyplot as plt 
plt.figure()
plt.hist(count,bins=100);
# plt.hist(count,bins=100,range=(0,2000));
plt.title("Histogram of Authors with given Article Counts");
plt.xlabel("Log$_{10}$( 1 + Article Number )");
plt.ylabel("Number of Authors");
# plt.savefig("/Users/jvenderley/Desktop/muck_rack-hist.pdf")
```
